```python
import os
import json

class Gasolineria:

    def __init__(self):
        self.corridas_archivo = "corrida.json"
        self.contador_archivo = "contador.txt"
        self.capacidad = 3.785
        self.precio_lt = 8.20
        self.cant_galones = 0
        self.cant_tg = 0
        self.apagar = 0

    def gas(self):
        print("La capacidad de cada galón es de 3.785 lt")
        print("El precio es de $8.20 por litro")

    def total(self):
        self.cant_galones = int(input("¿Cuántos galones necesitas?:"))
        self.cant_tg = self.cant_galones * self.capacidad
        self.apagar = self.cant_tg * self.precio_lt

    def imprimir(self):
        print("El total de gasolina a cargar es:", self.cant_tg, "lt")
        print("El precio total a pagar es de: $", self.apagar)

    def registrar_ejecucion(self):
        if os.path.exists(self.contador_archivo):
            with open(self.contador_archivo, "r") as file:
                contador = int(file.read())
        else:

            contador = 0


        contador += 1


        print(f"corrida {contador} .")


        with open(self.contador_archivo, "w") as file:
            file.write(str(contador))


        registro_actual = {
            "corrida": contador,
            "total_gasolina": self.cant_tg,
            "total_pagar": self.apagar
        }

        registros = []


        if os.path.exists(self.corridas_archivo):
            with open(self.corridas_archivo, "r") as file:
                registros = json.load(file)


        registros.append(registro_actual)


        with open(self.corridas_archivo, "w") as file:
            json.dump(registros, file, indent=2)


if __name__ == "__main__":
    cliente1 = Gasolineria()
    cliente1.gas()
    cliente1.total()
    cliente1.registrar_ejecucion()
    cliente1.imprimir()
```
